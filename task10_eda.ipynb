```python
Q1 = df['Exam_Score'].quantile(0.25)
Q3 = df['Exam_Score'].quantile(0.75)
IQR = Q3 - Q1 
Q1, Q3, IQR
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
lower_bound, upper_bound
```

```python
outliers = df[(df['Exam_Score'] < lower_bound) | (df['Exam_Score'] > upper_bound)]
outliers.shape
```

```python
outliers.head()
```

```python
df['ExamScore_Outlier'] = (
    (df['Exam_Score'] < lower_bound) | (df['Exam_Score'] > upper_bound)
)
```

```python
df['ExamScore_Outlier'].value_counts()
```

```python
df_cleaned = df[~df['ExamScore_Outlier']]
```

```python
df.shape, df_cleaned.shape
```

```python
corr_matrix = df_cleaned.corr(numeric_only=True)
corr_matrix
```

```python
corr_matrix['Exam_Score'].sort_values(ascending=False)
```
